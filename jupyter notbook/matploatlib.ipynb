```python
import numpy as np
import matplotlib.pyplot as plt

x = ["Python", "C", "C++", "Java"]
y = [85, 70, 60, 82]
z = [20, 30, 40, 50]
width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("Language", fontsize=15)
plt.ylabel("Popularity (%)", fontsize=15)
plt.title("Popularity Analysis")

c = ["y", "b", "m", "g"]
plt.bar(p, y, width, color="r", label="Current Popularity")
plt.bar(p1, z, width, color="b", label="Previous Popularity")

plt.xticks(p + width / 2, x)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

day = [1, 2, 3, 4, 5, 6, 7]
no = [50, 30, 40, 20, 10, 80, 100]
sizes = [10, 20, 30, 40, 50, 60, 70]
color = [10, 20, 30, 40, 50, 60, 70]

plt.scatter(day, no, c=color, s=sizes, cmap="BrBG")
plt.title("Scatter plot", fontsize=15)
t = plt.colorbar()
t.set_label("Colorbar", fontsize=15)
plt.show()
```

#Pie ploat

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40]
y = ["C", "C++", "Java", "Python"]

plt.pie(x, labels=y, startangle=90, shadow=True, explode=(0, 0, 0, 0.1), autopct='%1.1f%%', radius=1, textprops={"fontsize" : 13, "color" : "black"})
plt.show()
```

```python
plt.pie([1])
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10, 20, 30, 40]
x1 = [20, 30, 40, 50]
y = ["C", "C++", "Java", "Python"]

plt.pie(x, labels=y, radius=1, autopct='%1.1f%%')
plt.pie(x1, labels=y, radius=.5)
plt.show()
```

Stem Plot

```python
x = [1, 2, 3, 4, 5, 6]
y = [1, 2, 1, 3, 2, 4]

plt.stem(x,y, linefmt=":", markerfmt="ro", basefmt="-.",bottom=1.5, label = "Stem plot", )
plt.legend()
plt.show()
```

Box plot

```python
import matplotlib.pyplot as plt

x= [1,2,3,4,5,6,7]
plt.boxplot(x, widths=.2, labels=["Python"], patch_artist=True, showmeans=True, boxprops = dict(color = "r"), capprops=dict(color = "g"))

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,7]
y = [1,4,7,2,1,6,4]
z = [x,y]
plt.boxplot(z, widths=.2, labels=["Python", "Java"], patch_artist=True, showmeans=True, boxprops = dict(color = "r"), capprops=dict(color = "g"))

plt.show()
```

Stack Plot

```python
x = [1,2,3,4,5,6,7]
area = [1,4,7,2,1,6,4]
area2 = [2,5,8,3,2,7,5]
area3 = [3,6,9,4,3,8,6]

l= ["Python", "Java", "C++"]
plt.stackplot(x,area,area2,area3, labels= l, colors = ["r","g","b"])
plt.title("data")
plt.legend()
plt.show()


```

Step plot

```python
x = [1,2,3,4,5,6,7]
y = [1,4,7,2,1,6,4]

plt.step(x,y, marker = "o", where = "mid", mfc = "r")
plt.grid()
plt.show()
```

Fill_Between Plot

```python
x= [1,2,3,4,5,6,7]
y= [1,4,7,2,1,6,4]

plt.plot(x,y,color="r")
plt.fill_between(x,y, color="r", alpha=.5)
plt.show()
```

Sub plot

```python
x= [1,2,3,4,5,6,7]
y= [1,4,7,2,1,6,4]

plt.subplot(2,2,1)
plt.plot(x,y,color="r")

plt.subplot(2,2,2)
plt.pie([1])

x1= [1,2,3,8,5,4,7]
plt.subplot(2,2,3)
plt.pie(x1)

x2 = ["a", "b", "c", "d", "e", "f", "g"]
y2 = [1,4,7,2,1,6,4]
plt.subplot(2,2,4)
plt.bar(x2,y2)

plt.show()
```
