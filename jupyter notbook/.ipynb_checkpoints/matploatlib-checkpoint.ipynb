```python
import numpy as np
import matplotlib.pyplot as plt

x = ["Python", "C", "C++", "Java"]
y = [85, 70, 60, 82]
z = [20, 30, 40, 50]
width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("Language", fontsize=15)
plt.ylabel("Popularity (%)", fontsize=15)
plt.title("Popularity Analysis")

c = ["y", "b", "m", "g"]
plt.bar(p, y, width, color="r", label="Current Popularity")
plt.bar(p1, z, width, color="b", label="Previous Popularity")

plt.xticks(p + width / 2, x)
plt.legend()
plt.show()
```

Scatter plot

```python
import matplotlib.pyplot as plt
```

```python
day = [1,2,3,4,5,6,7]
no = [2,1,3,4,7,6,5]

sizes = [10,2,7,9,9,20,100]

plt.scatter(day, no, color = "r", s = sizes)
plt.title("Scatter ploat", fontsize = 15 )

plt.xlabel("Day", fontsize = 15)
plt.ylabel("no", fontsize = 15)
plt.show()
```
